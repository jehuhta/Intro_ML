```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Phone data as per lecture materials.
# x = screen time in minutes
# y = percent battery used
phones = {
    "x":[568,341,609,477,511],
    "y":[75,46,87,59,61]
}

df = pd.DataFrame(phones)
df
```

```python
# First, let's retrieve the means of the x and y columns.
x_mean = df['x'].mean()
y_mean = df['y'].mean()

# This point will always pass through the regression line.
print(x_mean, y_mean)
```

```python
# Lets find the distances of each point from the means
df['x_diff'] = df['x'] - x_mean
df['y_diff'] = df['y'] - y_mean
df.head(3)
```

```python
df['x_diff * y_diff'] = df['x_diff'] * df['y_diff']
df["x_diff^2"] = df["x_diff"] ** 2
df.head(2)
```

```python
numerator = df['x_diff * y_diff'].sum()
denomenator = df['x_diff^2'].sum()

m = numerator / denomenator
print(m)
```

```python
b = y_mean - m * x_mean
print(b.round(2))
```

```python
line_regression = f"y = {m.round(2)}x + {b.round(2)}"
line_regression
```
